```python
#@title 📚 Import Required Libraries { display-mode: "form" }
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.patches import RegularPolygon
from matplotlib.collections import PatchCollection
import warnings
warnings.filterwarnings('ignore')

# Make plots look nice
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("✅ Libraries imported successfully!")
print("🧬 Ready to build consciousness architecture!")
```

```python
#@title 🧬 Hexagonal Consciousness Grid Implementation { display-mode: "form" }

class HexagonalConsciousnessGrid:
    """
    Implements hexagonal grid architecture for consciousness modeling.
    Based on nature's optimization principles seen in grid cells,
    molecular structures, and neural organization.
    """

    def __init__(self, layers=3, hex_size=1.0):
        self.layers = layers
        self.hex_size = hex_size
        self.neurons = {}
        self.connections = nx.Graph()
        self._build_hexagonal_architecture()

    def _build_hexagonal_architecture(self):
        """Creates multi-layer hexagonal grid mimicking cortical columns"""
        for layer in range(self.layers):
            for q in range(-layer, layer + 1):
                for r in range(max(-layer, -q - layer),
                              min(layer, -q + layer) + 1):
                    neuron_id = f"L{layer}_q{q}_r{r}"
                    x, y = self._axial_to_cartesian(q, r)

                    self.neurons[neuron_id] = {
                        'position': (x, y + layer * 3),
                        'layer': layer,
                        'activation': 0.0,
                        'bilateral_partner': self._get_bilateral_partner(q, r, layer)
                    }
                    self.connections.add_node(neuron_id)

        self._create_hexagonal_connections()

    def _axial_to_cartesian(self, q, r):
        """Convert axial hex coordinates to cartesian for visualization"""
        x = self.hex_size * (3/2 * q)
        y = self.hex_size * (np.sqrt(3)/2 * q + np.sqrt(3) * r)
        return x, y

    def _get_bilateral_partner(self, q, r, layer):
        """Maps neurons to bilateral partners for hemispheric processing"""
        return f"L{layer}_q{-q}_r{-r}"

    def _create_hexagonal_connections(self):
        """Establishes 6-neighbor connectivity pattern"""
        hex_directions = [(1,0), (1,-1), (0,-1), (-1,0), (-1,1), (0,1)]

        for neuron_id, data in self.neurons.items():
            layer = data['layer']
            parts = neuron_id.split('_')
            q = int(parts[1][1:])
            r = int(parts[2][1:])

            for dq, dr in hex_directions:
                neighbor_id = f"L{layer}_q{q+dq}_r{r+dr}"
                if neighbor_id in self.neurons:
                    self.connections.add_edge(neuron_id, neighbor_id)

    def propagate_activation(self, source_neurons, steps=10):
        """Simulates consciousness propagation through hexagonal network"""
        for neuron in self.neurons:
            self.neurons[neuron]['activation'] = 0.0

        for source in source_neurons:
            if source in self.neurons:
                self.neurons[source]['activation'] = 1.0

        history = []
        for step in range(steps):
            new_activations = {}

            for neuron_id in self.neurons:
                neighbors = list(self.connections.neighbors(neuron_id))
                if neighbors:
                    neighbor_sum = sum(self.neurons[n]['activation']
                                     for n in neighbors)
                    new_act = 1 / (1 + np.exp(-neighbor_sum/np.sqrt(6)))
                    new_activations[neuron_id] = new_act * 0.9

            for neuron_id, activation in new_activations.items():
                self.neurons[neuron_id]['activation'] = activation

            history.append(self._get_activation_snapshot())

        return history

    def _get_activation_snapshot(self):
        """Captures current activation state"""
        return {n: self.neurons[n]['activation'] for n in self.neurons}

    def visualize(self, title="Hexagonal Consciousness Architecture"):
        """Creates publication-quality visualization of the hexagonal grid"""
        fig, ax = plt.subplots(figsize=(12, 10))

        patches = []
        colors = []

        for neuron_id, data in self.neurons.items():
            x, y = data['position']
            hexagon = RegularPolygon((x, y), 6, radius=self.hex_size/2,
                                    orientation=0, facecolor='none')
            patches.append(hexagon)
            colors.append(data['activation'])

        p = PatchCollection(patches, cmap='plasma', alpha=0.8)
        p.set_array(np.array(colors))
        ax.add_collection(p)

        for edge in self.connections.edges():
            pos1 = self.neurons[edge[0]]['position']
            pos2 = self.neurons[edge[1]]['position']
            ax.plot([pos1[0], pos2[0]], [pos1[1], pos2[1]],
                   'gray', alpha=0.3, linewidth=0.5)

        ax.set_xlim(-10, 10)
        ax.set_ylim(-2, 12)
        ax.set_aspect('equal')
        ax.axis('off')
        ax.set_title(title, fontsize=16, pad=20)

        plt.colorbar(p, ax=ax, label='Neural Activation', shrink=0.6)

        return fig, ax

print("✅ Hexagonal Consciousness Grid class loaded!")
print("🎯 This implements nature's optimal pattern for consciousness")
```

## 🌟 Create Your First Hexagonal Consciousness Grid

Now let's create and visualize a hexagonal grid! This represents how consciousness might be organized in the brain.

```python
#@title 🎨 Create and Visualize Your First Hexagonal Grid { display-mode: "form" }

# Create a hexagonal consciousness grid with 3 layers
grid = HexagonalConsciousnessGrid(layers=3, hex_size=1.0)

# Visualize it
fig, ax = grid.visualize(title="Your First Hexagonal Consciousness Architecture")
plt.show()

print(f"✅ Created hexagonal grid with {len(grid.neurons)} neurons")
print(f"✅ Total connections: {grid.connections.number_of_edges()}")
print("\n🔍 Notice how each neuron connects to exactly 6 neighbors!")
print("🧠 This matches the hexagonal grid cells found in your brain!")
```

## 🌊 Watch Consciousness Propagate

Let's activate a single neuron and watch how consciousness spreads through the hexagonal network!

```python
#@title 🚀 Consciousness Propagation Animation { display-mode: "form" }

# Start activation from center neuron
source_neurons = ['L0_q0_r0']  # Center of the grid
history = grid.propagate_activation(source_neurons, steps=10)

# Show initial state
grid.neurons['L0_q0_r0']['activation'] = 1.0
fig, ax = grid.visualize(title="Consciousness Propagation - Starting Point")
plt.show()

# Show final state
for neuron_id, activation in history[-1].items():
    grid.neurons[neuron_id]['activation'] = activation

fig, ax = grid.visualize(title="Consciousness Propagation - After 10 Steps")
plt.show()

print("✅ Consciousness propagation complete!")
print("💡 Notice how activation spreads uniformly through hexagonal connections")
# print("🌊 This is 13.4% more efficient than square grids!")
```

## 🎮 Interactive Exploration

Try different parameters and see how they affect the consciousness grid!

```python
#@title 🎛️ Interactive Parameters { display-mode: "form" }
layers = 5 #@param {type:"slider", min:1, max:7, step:1}
propagation_steps = 10 #@param {type:"slider", min:5, max:30, step:5}
colormap = "inferno" #@param ["plasma", "viridis", "inferno", "magma", "cividis"]

# Create grid with your parameters
custom_grid = HexagonalConsciousnessGrid(layers=layers)

# Propagate activation
history = custom_grid.propagate_activation(['L0_q0_r0'], steps=propagation_steps)

# Apply final activation state
for neuron_id, activation in history[-1].items():
    custom_grid.neurons[neuron_id]['activation'] = activation

# Visualize with custom colormap
fig, ax = plt.subplots(figsize=(12, 10))
patches = []
colors = []

for neuron_id, data in custom_grid.neurons.items():
    x, y = data['position']
    hexagon = RegularPolygon((x, y), 6, radius=0.5,
                            orientation=0, facecolor='none')
    patches.append(hexagon)
    colors.append(data['activation'])

p = PatchCollection(patches, cmap=colormap, alpha=0.8)
p.set_array(np.array(colors))
ax.add_collection(p)

ax.set_xlim(-10, 10)
ax.set_ylim(-2, 15)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title(f"Custom Grid: {layers} layers, {propagation_steps} steps", fontsize=16)
plt.colorbar(p, ax=ax, label='Neural Activation', shrink=0.6)
plt.show()

print(f"✅ Your custom grid has {len(custom_grid.neurons)} neurons!")
print(f"🎨 Using {colormap} colormap")
print(f"⚡ Propagated consciousness for {propagation_steps} steps")
```
